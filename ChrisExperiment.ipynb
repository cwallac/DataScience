```python
latValues = dict((el,0) for el in lats)
longValues = dict((el,0) for el in longs)
```

```python
for i in range(len(lats)):
    landVal = float(df.iloc[[i]].AV_LAND.values)
    latValues[lats[i]] += landVal
    longValues[longs[i]] += landVal
    
```

```python
latCdf = thinkstats2.MakeCdfFromDict(latValues)
longCdf = thinkstats2.MakeCdfFromDict(longValues)

thinkplot.PrePlot(cols=2)

thinkplot.SubPlot(1)
thinkplot.Cdf(latCdf)
thinkplot.Config(title='CDF Latitudes')

thinkplot.SubPlot(2)
thinkplot.Cdf(longCdf)
thinkplot.Config(title='CDF Longitudes')
```
