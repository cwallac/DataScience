```python
import numpy as np
import pandas as p 
import thinkstats2
import thinkplot
```

```python
df = p.DataFrame.from_csv("Property_Assessment_2014_O.csv")
```

```python
#df = totaldf[totaldf.AV_LAND > 0]
TaxDf = df[df.AV_LAND > 0]
cdf = thinkstats2.Cdf(df.GROSS_TAX)
thinkplot.Cdf(cdf)
thinkplot.Config(xscale="log",title="Tax collected from properties",xlabel="Dollars",ylabel="Percentage")
```

```python
#df = totaldf[totaldf.AV_LAND > 0]
cdf = thinkstats2.Cdf(df.LAND_SF)
thinkplot.Cdf(cdf)
thinkplot.Config(xscale="log",title="Property Footage CDF",xlabel="Square Feet",ylabel="Percentage")
```

```python
#df = totaldf[totaldf.AV_LAND > 0]
LandDf = df[df.AV_LAND > 0]
cdf = thinkstats2.Cdf(LandDf.AV_LAND)
thinkplot.Cdf(cdf)
thinkplot.Config(xscale="log",title="Land Value CDF",xlabel="Dollars",ylabel="Percentage")
```

```python
#df = totaldf[totaldf.AV_LAND > 0]
BldgDf = df[df.AV_BLDG > 0]
cdf = thinkstats2.Cdf(BldgDf.AV_BLDG)
thinkplot.Cdf(cdf)
thinkplot.Config(xscale="log",title="Building Value CDF",xlabel="Dollars",ylabel="Percentage")
```

```python
yeardf = df[df.YR_BUILT > 0]
cdf = thinkstats2.Cdf(yeardf.YR_BUILT)
thinkplot.Cdf(cdf)
thinkplot.Config(title="Year Built",xlabel="Year",ylabel="Percentage")
```

```python
year = yeardf.YR_BUILT
value = yeardf.AV_BLDG
thinkplot.Scatter(year,value,alpha=.2)
#thinkplot.HexBin(others_footage,others_values)
thinkplot.Show(label="blarg",xlabel="Footage",yscale="log",ylabel="Value",xlim=[1900,2050],ylim=[0,100000000])
```
