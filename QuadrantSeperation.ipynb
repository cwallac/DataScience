```python
import numpy as np
import pandas as p 
import thinkstats2
import thinkplot
import bisect
```

```python
df = p.DataFrame.from_csv("Property_Assessment_2014_O.csv")
```

```python
#GEt LAt and Lng into functioning values
lats = [0] * len(df.Location.values)
longs = [0] * len(df.Location.values)
for i in range(len(df.Location.values)):
    latLong = df.Location.values[i].strip("()").split(",")
    try:
        lats[i] = float(latLong[0])
        longs[i] = float(latLong[1])
    except: # Data error
        print i # THis should be the excel row -2
        latLong = df.Location.values[i-1].strip("()").split(",")
        lats[i] = float(latLong[0])
        longs[i] = float(latLong[1])
#Binning the values        
resolution = 50
latLookup = np.linspace(min(lats),max(lats),resolution)
longLookup = np.linspace(min(longs),max(longs),resolution)

quadrantArray = [0] * len(df.Location.values)
for i in range(len(lats)):
    landVal = float(df.iloc[[i]].AV_LAND.values)
    latIndex = bisect.bisect_left(latLookup,lats[i])
    longIndex = bisect.bisect_left(longLookup,longs[i])
    quadrantArray[i] = resolution * latIndex+longIndex
    
df["QUADRANT"] = quadrantArray #Easy way to access bins, increase from left to right and down from 0 to 2499

#GEt Values of 
numberProps = dict((el,0) for el in range(len(quadrantArray))) #This tells us # of entries in teh quadrant
for i in range(len(quadrantArray)):
    numberProps[i] = len(df[df.QUADRANT == i])
```
