```python
import pandas as p
df = p.DataFrame.from_csv("QuadrantData.csv")
df
```

```python
import thinkstats2
import thinkplot
df = df[df.NUM_PROPS > 0]
crimes = df.NUM_CRIMES
cdf = thinkstats2.Cdf(crimes)
thinkplot.Cdf(cdf)
thinkplot.config(title="Crimes commited per zone",xlabel="Crimes since 2012",xscale="log")
```

```python
import statsmodels.formula.api as smf
df = df[df.NUM_PEOPLE > 0]

formula = "CRIME_RATE ~ NUM_PEOPLE + DIST_STATION"
model = smf.ols(formula,data=df)
results = model.fit()
```
